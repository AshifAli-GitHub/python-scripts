```python
from langgraph.graph import StateGraph,START,END
#from langchain_openai import OpenAI 
from typing import TypedDict
#from dotenv import load_dotenv

```

```python
class batsman_state(TypedDict):
    balls:int
    runs:int 
    sixes:int
    fours:int 
    Strike_rate:float
    boundary_run_percentage:float
    balls_per_boundary:float
    summary:str
    
```

```python
def calculate_strike_rate_python(state:batsman_state)->batsman_state:
    balls=state['balls']
    runs=state['runs']

    strike_rate= (runs/balls)*100
    state['Strike_rate']=strike_rate
    return {'Strike_rate':strike_rate}
```

```python
def calculate_boundary_run_percentage_python (state:batsman_state)-> batsman_state:
    sixes=state['sixes']
    fours=state['fours']
    runs=state['runs']
    boundary_run=(sixes*6)+(fours*4)
    boundary_run_percentage=(boundary_run/runs)*100
    state['boundary_run_percentage']=boundary_run_percentage

    return {'boundary_run_percentage':boundary_run_percentage}

```

```python
def calculate_balls_per_boundary_python (state:batsman_state)->batsman_state:
    sixes=state['sixes']
    fours=state['fours']
    balls=state['balls']
    no_of_boundaries=sixes+fours
    balls_per_boundary=balls/no_of_boundaries

  #  balls_per_boundary=state['balls']/(state['sixes']+state['fours'])
    state['balls_per_boundary']=balls_per_boundary

    return {'balls_per_boundary': balls_per_boundary}
```

```python
def summary_python (state:batsman_state)->batsman_state:
    summary=f"""
    strike rate - {state['Strike_rate']}\n
    balls per boundary - {state['balls_per_boundary']}\n
    boundary percentage - {state['boundary_run_percentage']}\n
    """
    state['summary']=summary
    return state
```

```python
graph=StateGraph(batsman_state)

graph.add_node('calculate_strike_rate',calculate_strike_rate_python)
graph.add_node('calculate_boundary_run_percentage',calculate_boundary_run_percentage_python)
graph.add_node('calculate_balls_per_boundary',calculate_balls_per_boundary_python)
graph.add_node('summary',summary_python)

graph.add_edge(START,'calculate_strike_rate')
graph.add_edge(START,'calculate_boundary_run_percentage')
graph.add_edge(START,'calculate_balls_per_boundary')
graph.add_edge('calculate_strike_rate','summary')
graph.add_edge('calculate_boundary_run_percentage','summary')
graph.add_edge('calculate_balls_per_boundary','summary')

graph.add_edge('summary',END)

workflow=graph.compile()





```

```python
initial_state={'balls':80,'runs':60,'sixes':3,'fours':5}
final_state=workflow.invoke(initial_state)
print(final_state)
```
